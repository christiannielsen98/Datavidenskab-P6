```python
def json_to_dataframe(json_file):
    if "," in json_file[0]["name_node"]:
        level_df = pd.DataFrame(columns=["pattern","supp","conf", "time"])
        for i in json_file:
            for j in i["patterns"]:
                level_df.loc[level_df.shape[0]] = j
    else:
        level_df = pd.DataFrame(columns=["name_node","supp","conf"])
        for i in json_file:
            level_df.loc[level_df.shape[0]] = i
    return level_df
```

```python
def avg_event_pr_day(max_day, dictionary):
    list = []
    for values in dictionary.values():
        list.append(len(values))
    return sum(list)/max_day
```

```python
max_day = -1
for index, row in level2_df.iterrows():
    max_day = max(max_day, max(int(key) + 1 for key in row["time"].keys()))

for index, row in level2_df.iterrows():
    level2_df.loc[index, "avg_event_pr_day"] = avg_event_pr_day(max_day, row["time"])
```

```python
def start_end_times_of_rules(dictionary):
    list = []
    for values in dictionary.values():
        for i in values:
            start_time = int(i[0:1][0][0:1][0].split(":")[0].split(" ")[1])
            end_time = int(i[1:2][0][1:2][0].split(":")[0].split(" ")[1])
            list.append([start_time] + [end_time])
    temp_df = pd.DataFrame(list, columns = ['start_time', 'end_time'])
    start_end_times_df = temp_df.groupby(temp_df.columns.tolist(),as_index=False).size()
    return start_end_times_df
```

```python
def SE_time_df(dataframe=level2_df):
    df = pd.DataFrame(columns=['rule', 'start_time', 'end_time', 'size'])
    for index, row in dataframe.iterrows():
        # level2_df.loc[index, "avg_event_pr_day"] = avg_event_pr_day(row["time"])

        temporary_df = start_end_times_of_rules(row["time"])
        temporary_df["rule"] = row["pattern"]
        df = pd.concat((df, temporary_df), ignore_index=True)
    return df#.loc[df["size"]>52]
SE_time_df = SE_time_df()
SE_time_df
```
