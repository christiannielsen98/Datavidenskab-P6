```python
def json_to_dataframe(json_file):
    if "," in json_file[0]["name_node"]:
        level_df = pd.DataFrame(columns=["pattern","supp","conf", "time"])
        for i in json_file:
            for j in i["patterns"]:
                level_df.loc[level_df.shape[0]] = j
    else:
        level_df = pd.DataFrame(columns=["name_node","supp","conf"])
        for i in json_file:
            level_df.loc[level_df.shape[0]] = i
    return level_df
```

# Counting appliance occurences in level 2 rules

```python
from collections import Counter


follows_left_list = []
follows_right_list = []
contains_left_list = []
contains_right_list = []
overlaps_left_list = []
overlaps_right_list = []

for i in level2_df["pattern"].index:
    rule = level2_df["pattern"][i]
    if "->" in rule:
        appliance_left = rule.split("->")[0]
        follows_left_list.append(appliance_left)
        appliance_right = rule.split("->")[1]
        follows_right_list.append(appliance_right)
    elif ">" in rule:
        appliance_left = rule.split(">")[0]
        contains_left_list.append(appliance_left)
        appliance_right = rule.split(">")[1]
        contains_right_list.append(appliance_right)
    elif "|" in rule:
        appliance_left = rule.split("|")[0]
        overlaps_left_list.append(appliance_left)
        appliance_right = rule.split("|")[1]
        overlaps_right_list.append(appliance_right)
```

```python
dict_of_counts_level2 = {"follows_left": follows_left_list, "follows_right": follows_right_list, "contains_left": contains_left_list, "contains_right": contains_right_list, "overlaps_left": overlaps_left_list, "overlaps_right": overlaps_right_list}

for key, values in dict_of_counts_level2.items():
    dict_of_counts_level2[key] = dict(Counter(values))

dict_of_counts_level2
```

```python
level2_df_stats = pd.DataFrame(dict_of_counts_level2).fillna(0)
level2_df_stats.sort_values(by=["follows_left"], ascending=False, inplace=True)
level2_df_stats
```

```python
fig = px.bar(level2_df_stats)


fig.update_layout(margin=dict(
    l=30,
    r=30,
    b=30,
    t=50,
), xaxis_nticks=30,
    # annotations=annotations,
)
fig.update_layout(barmode='group')
fig.show()
```
